8. Let's deal with the missing values. 

a) For the features EthnicGroup, ParentMaritalStatus and WklyStudyHours replace NaN values with the most common value.

b) For the feature ParentEduc replace NaN value with a string *no formal education*.

c) For the feature TestPrep replace NaN value with a string *none*.

d) For the features PracticeSport and IsFirstChild remove all the examples with NaN values. 

e) For the feature NrSiblings replace the NaN value with the median value.

f) Finally, for the feature TransportMeans replace the NaN value with a string *walking*.




```python
#a)
features_most_common = ['EthnicGroup', 'ParentMaritalStatus', 'WklyStudyHours']
for feature in features_most_common:
    most_common_value = students[feature].mode()[0]
    students[feature].fillna(most_common_value, inplace=True)

#b)
students['ParentEduc'].fillna('no formal education', inplace=True)

#c)
students['TestPrep'].fillna('none', inplace=True)

#d)
students.dropna(subset=['PracticeSport', 'IsFirstChild'], inplace=True)

#e)
nrsiblings_median = students['NrSiblings'].median()
students['NrSiblings'].fillna(nrsiblings_median, inplace=True)

#f)
students['TransportMeans'].fillna('walking', inplace=True)

missing_values_check = students[['EthnicGroup', 'ParentMaritalStatus', 'WklyStudyHours', 'ParentEduc', 'TestPrep', 'PracticeSport', 'IsFirstChild', 'NrSiblings', 'TransportMeans']].isnull().sum()
print(missing_values_check)
```

9. Are there inconsistencies in the IsFirstChild column? Replace the inconsistent entries so they match the rest of the values. 

```python
students['IsFirstChild'].unique()
```

```python
students['IsFirstChild'] = students['IsFirstChild'].replace(['ye', 'si','Yes', 'yeah', 'ys', 'yeah boy', 'es'], 'yes')
students['IsFirstChild'] = students['IsFirstChild'].replace(['nein', 'nope', 'noo'], 'no')

students['IsFirstChild'].unique()
```

10. The final step to preparing the dataset for classification is transforming the non-numeric variables to numeric variables. 
For features Gender and LunchType use OneHot Encoder, and for EthnicGroup, ParentEduc, TestPrep, ParentMaritalStatus, PracticeSport, IsFirstChild, TransportMeans and WklyStudyHours use Label Encoder. 

```python
encoder_gender = OneHotEncoder(handle_unknown="ignore")
students_tmp_gender = encoder_gender.fit_transform(students[['Gender']])
students_tmp_gender = pd.DataFrame(students_tmp_gender.toarray(), columns=encoder_gender.get_feature_names_out(["Gender"]))
students = pd.concat([students.reset_index(drop=True), students_tmp_gender], axis=1)
students.drop(['Gender'], axis=1, inplace=True)

encoder_lunch = OneHotEncoder(handle_unknown="ignore")
students_tmp_lunch = encoder_lunch.fit_transform(students[['LunchType']])
students_tmp_lunch = pd.DataFrame(students_tmp_lunch.toarray(), columns=encoder_lunch.get_feature_names_out(['LunchType']))
students = pd.concat([students.reset_index(drop=True), students_tmp_lunch], axis=1)
students.drop(['LunchType'], axis=1, inplace=True)

students.head()
```

```python
label_encode_columns = ['EthnicGroup', 'ParentEduc', 'TestPrep', 'ParentMaritalStatus', 'PracticeSport', 'IsFirstChild', 'TransportMeans', 'WklyStudyHours']

label_encoder = LabelEncoder()
for column in label_encode_columns:
    students[column] = label_encoder.fit_transform(students[column])

students.head()
```
